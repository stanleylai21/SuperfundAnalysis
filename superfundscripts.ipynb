```python
import datetime
import re
def convert_to_epoch(date):
    #sample date 07/22/1987
    datelist = re.split("/", date)
    month = int(datelist[0])
    day = int(datelist[1])
    year = int(datelist[2])
    return datetime.datetime(year, month, day).timestamp()

convert_to_epoch('4/1/2012')
    
```

```python
def find_state(address):
    #sample address N FRANKLIN ST EXT, CHAMBERSBURG, PA 17201
    addresslist = re.split(",", address)
    relevant = addresslist[len(addresslist)-1]
    statezip = relevant.strip()[:2]
    return statezip
```

```python
shorterfund['state'] = shorterfund['address'].apply(find_state)
```

```python
contamslist = []
base_pickle = "C:\CS 122\Final Project\contamdumps"
for i in range(len(id_list)):
    picklename = base_pickle + "\site" + str(i) + "contams.p"
    favorite_color = pickle.load(open(picklename,"rb"))
    soup = BeautifulSoup(favorite_color, 'html.parser')
    divs = soup.findAll('tbody')
    try:
        contams = divs[0].findAll('tr')
        innerlist = []
        for conlist in contams:
            a = conlist.text
            b = re.split("\n", a.strip())
            innerlist.append([b[1], b[2]])
        contamslist.append(innerlist)
    except IndexError:
        contamslist.append("None")
print("Done")
```
